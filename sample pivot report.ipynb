```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import the data from CSV

```python
prices = pd.read_csv('price_paid_records.csv')
```

```python
prices
```

Pivot data indexed against county displaying median house price data in accending order 

```python
county = prices.pivot_table(values='Price',  index='County',  aggfunc='median')
county.sort_values(by=['Price'], ascending=True, inplace=True )
county
```

Display data as bar chart

```python
county.plot(kind = 'bar',figsize=(30, 20), rot=90)
```
